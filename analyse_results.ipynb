```python
import pandas as pd
import os
```

```python
files = os.listdir('measurements')
files = sorted([f for f in files if f.startswith('data2neo') or f.startswith('csv')])
files= [(f.split('_')[0],f.split('_')[1], f) for f in files]
```

```python
df = pd.DataFrame(columns=['algo', 'time', 'repo'])
for algo, repo, f in files:
    df2 = pd.read_csv('measurements/' + f, header=None, names=['time'])
    df2['repo'] = repo
    df2['algo'] = algo
    if 'csv' in f:
        # just first row (second row is without csv export)
        df2 = df2.iloc[:1]

    df = pd.concat([df, df2])
df.reset_index(drop=True, inplace=True)
df
```

```python
# create table with mean and std
results = df.groupby(['repo', 'algo']).agg(['mean', 'std'])
# convert to minutes
results['time'] = results['time'] / 60
results.T
```
